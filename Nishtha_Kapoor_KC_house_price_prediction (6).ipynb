# Handle missing values

```python
from sklearn.impute import SimpleImputer
sim=SimpleImputer(missing_values=np.nan,strategy="median")
X[:,[0]]=sim.fit_transform(X[:,[0]])
```

```python
X=pd.DataFrame(X)
X.isnull().sum()
```

# Applying Standard deviation Normalization


### Normalization usually means to scale a variable to have a values between 0 and 1 ..and this technique is useful when our data has varying scales

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
X
```

# Outliers

### outliers is an observation point that is distant from other observations and it occurs due to some experimental measurement error in the dataset or having variability in the data.. in some cases excluding outliers can cause our result to become statistically significant


# Discover  outliers  by using IQR score

# 1. calculating first quartile and third quartile range

```python
q1=dataset.quantile(0.25)
q3=dataset.quantile(0.75)
```

# 2. Find interquartile range

```python
iqr=q3-q1
print(iqr)
```

# 3. apply iqr score to remove outliers

```python
dataset=dataset[~((dataset<(q1-1.5*iqr))|(dataset>(q3+1.5*iqr))).any(axis=1)]
dataset.shape

```

# Add constant in our variable X

```python
import statsmodels.api as sn
X_cons=sn.add_constant(X)
X_cons
```

# Now we fit our model

```python
lm=sn.OLS(y,X_cons).fit()
lm.summary()
```

# Spliting the data into training set and test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

# Implementing Linear Regression

### In Regression,the output variable is a continous numeric value. The assumption of this algorithm is that the data must be linearly distributed.The simple linear regression attempts to find a line(hyperplane)that can explain the trend present in the dataset in an optimal fashion.

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
```

```python
print(linreg.intercept_,linreg.coef_)
```

# Train the model by using training sets

```python
linreg.fit(X_train,y_train)
```

# Prediction on the Training Data

```python
y_train_pred=linreg.predict(X_train)
y_train_pred
```

# Evaluating Model's Performance on Training set

## 1. R-Squared

### The r-square is known as coefficient of determination.This value is basically a ratio that always outputs a value b/w the limits of [0,1].A closer to 1 value specifies good behaviour of the model and also a closer to 0  value specifies poor model.



```python
from sklearn.metrics import r2_score
r=r2_score(y_train,y_train_pred)
print('R^2:',r)
```

## 2. Adj. R-Squared

### The Adj. r-square metric has been smartly derived to punish the model for adding irrelevant features by lowering its value.Moreover,it also rewards the model for adding relevant features or strong predictors.

```python
print('Adjusted R^2:',1 - (1-r)*(len(y_train)-1)/(len(y_train)-X_train.shape[1]-1))
```

## 3. Root mean square error

### RMSE is a standard way to measure the error of a model in predicting continous data. 

```python
from sklearn.metrics import mean_squared_error
print('RMSE:',np.sqrt(mean_squared_error(y_train,y_train_pred)))
```

## 4. Mean Absolute Error

### This measures the absolute average distance between the Actual observation  and the predicted value

```python
from sklearn.metrics import mean_squared_error
print('MAE:',mean_squared_error(y_train,y_train_pred))
```
